```python
class MyHTMLParser(HTMLParser):
    def __init__(self):
        HTMLParser.__init__(self)
        self.twitterFlag = False
        self.data = []
    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            for (x, url) in attrs:
                if x == 'href' and 'https://www.twitter.com/' in url:
                    self.data.append(url)
```

```python
def get_twitter(url):
  try:
    html = requests.get(url)
    parser = MyHTMLParser()
    parser.feed(html.text)
    return parser.data
  except:
    print(url)
    return []
```
